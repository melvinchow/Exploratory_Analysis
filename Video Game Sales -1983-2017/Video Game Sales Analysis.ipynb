# Video Game Sales Analysis (1983-2017)

Data from: https://data.world/julienf/video-games-global-sales-in-volume-1983-2017

Numbers in Millions of Copies sold to date

```python
import pandas as pd
```

```python
df = pd.read_csv('vgsalesGlobale.csv', index_col=0)
```

```python
df.head(5)
```

```python
df.columns.tolist()
```

# Combine titles by Platform
to see sales figures independent of platform

```python
noPlatforms = df.groupby(['Name', 'Publisher'],as_index=False).agg({'Genre' : 'min','Year' : 'min',
                      'NA_Sales' : 'sum', 
                      'EU_Sales' : 'sum', 
                      'JP_Sales' : 'sum', 
                   'Other_Sales' : 'sum', 
                  'Global_Sales' : 'sum'})
```

```python
noPlatforms[['Name','Publisher','Genre','Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
           ].sort_values('Global_Sales', ascending = False).head(10)
```

# Best Racing Games

```python
noPlatforms[noPlatforms['Genre'] == 'Racing'].sort_values('Global_Sales', ascending = False).head(15)
```

# Racing Games: Forza

```python
noPlatforms[noPlatforms['Name'].str.contains('Forza')].sort_values('Global_Sales', ascending = False)
```

# Racing Games: Mario Kart

```python
noPlatforms[noPlatforms['Name'].str.contains('Mario Kart')].sort_values('Global_Sales', ascending = False)
```

# Ubisoft Best Sellers

Sorted by Global Sales

```python
Ubisoft = noPlatforms[noPlatforms['Publisher'] == 'Ubisoft']
Ubisoft.sort_values('Global_Sales', ascending=False).head(18)
```

# Tom Clancy Title Sales (Platform Independent)

```python
# Tom Clancy titles only
Ubisoft_TomClancy = noPlatforms[(noPlatforms['Publisher'] == 'Ubisoft') & (noPlatforms['Name'].str.contains('Tom'))]

# Sort by Global sales
Ubisoft_TomClancy[['Name','Publisher','Global_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Year']
           ].sort_values('Global_Sales', ascending = False).head(15)
```

# Best Selling Genres

```python
BestGenres = df.groupby('Genre').sum()
BestGenres.drop('Year', axis=1, inplace=True)
```

```python
BestGenres[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
          ].sort_values(['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], ascending=False)
```

# Most Popular Platform

```python
ByPlatform = df.groupby('Platform').sum()
ByPlatform.drop('Year', axis=1, inplace=True)
ByPlatform.sort_values(['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], ascending=False)
```
